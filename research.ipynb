```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from load_matrices import load_test_matrices
```

```python
matrices = load_test_matrices()
```

```python
def visualize_matrix_structure(matrices):
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    axes = axes.flatten()
    
    for idx, (name, data) in enumerate(list(matrices.items())[:6]):
        A = data['A']
        
        if A.shape[0] < 5000:
            axes[idx].spy(A, markersize=0.1)
            axes[idx].set_title(f"{name}\n{A.shape[0]}×{A.shape[1]}, nnz={A.nnz}")
        else:
            axes[idx].text(0.5, 0.5, f"Too large\n{A.shape[0]}×{A.shape[1]}", 
                          ha='center', va='center')
            axes[idx].set_title(name)
    
    plt.tight_layout()
```

```python
visualize_matrix_structure(matrices)
```
